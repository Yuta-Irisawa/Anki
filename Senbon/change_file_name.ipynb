```python
def change_filename(foldername, identifyBegin, identifyLast, extention):
  filenameList = glob.glob('*')
  filenameDict = {}
  for filename in filenameList:
    index_num = int(filename[identifyBegin:identifyLast])
    filenameDict[index_num] = filename
  filenameDictSorted = sorted(filenameDict.items(), key=lambda x:x[0])
  for i,filename in enumerate(filenameDictSorted):
    newfilename = foldername + "_{:0=4}.{}".format((i+1), extention)
    os.rename(filename[1], newfilename)
```

```python
def rename_directory_content(directory_name, foldername):
  os.chdir(directory_name)
  os.chdir("./" + foldername)
  if foldername == "question" or foldername == "answer" or foldername == "reading_question" or foldername == "reading_answer":
    change_filename(foldername, -8, -4, "jpg")
  elif foldername == "audio" or foldername == "reading_audio":
    change_filename(foldername, 0, 3, "mp3")
```

audio か reading_audioを使って再度編集するときは、change_filename(foldername, -8, -4, "mp3")に変更すること
